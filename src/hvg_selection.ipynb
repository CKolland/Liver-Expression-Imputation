```python
import anndata as ad
import numpy as np
```

```python
adata = ad.read_h5ad("../data/test_data.h5ad", backed=True)
adata
```

```python
hvgs = adata.var[adata.var["highly_variable"]].copy()
hvgs
```

```python
hvgs_ranked = hvgs.sort_values("variances_norm", ascending=False)
hvgs_ranked
```

```python
genes_for_selection = hvgs_ranked[25:75]

# Choose random genes
np.random.seed(12062025)
selected_genes = np.random.choice(50, size=20, replace=False)

genes_to_exclude = genes_for_selection.iloc[selected_genes]
testing_mask = ~adata.var_names.isin(genes_to_exclude.index)
testing_genes = adata.var_names[testing_mask]
testing_genes
```

```python
testing_genes.to_series().to_csv("imputation_mask.csv", index=False, header=False)
```
